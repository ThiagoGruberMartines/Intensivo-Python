```python
import pandas as pd
import plotly.express as px


df = pd.read_csv("caso.csv")

display(df)
```

```python
# Deletando as colunas vazias
df = df.dropna(how="all", axis=1)

# Deletando as linhas vazias
df = df.dropna(how="any", axis=0)
```

```python
# Análise inicial
print(df["state"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Análise com gráficos
for coluna in df.columns:
    grafico = px.histogram(df, x=coluna, color="state", text_auto=True)
    
grafico.show()
```
